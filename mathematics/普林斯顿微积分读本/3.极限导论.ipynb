# 3.极限导论
## 3.6 夹逼定理
三明治定理又叫夹逼定理（Squeeze Theorem）
$$
\text{若} \quad g(x) \leq f(x) \leq h(x) \quad \text{且} \quad 
\lim\limits_{x \to a} g(x) = \lim\limits_{x \to a} h(x) = L
$$
$$
\text{则} \quad \lim\limits_{x \to a} f(x) = L
$$

举例说明：
$f(x) = x^2\sin(\frac{1}{x})$被$g(x)=x^2$和$g(x)=-x^2$夹逼因此：$\lim\limits_{x->0}{x^2\sin(1/x)} = 0 $
夹逼定理图像

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc('font', family='STHeiti')
# 定义 x 的范围，避开 x=0 以防止 sin(1/x) 计算问题
x = np.linspace(-0.1, 0.1, 400)
x = x[x != 0]  # 去掉 x=0

# 定义函数 f(x) = x^2 * sin(1/x) 及 夹逼函数 g(x) = x^2, h(x) = -x^2
y_f = x**2 * np.sin(1/x)
y_g = x**2
y_h = -x**2

# 绘制图像
plt.figure(figsize=(8, 6))
plt.plot(x, y_f, label=r'$f(x) = x^2 \sin(1/x)$', color='b')
plt.plot(x, y_g, label=r'$g(x) = x^2$', linestyle='dashed', color='r')
plt.plot(x, y_h, label=r'$h(x) = -x^2$', linestyle='dashed', color='g')
plt.legend()

# 图例与标题
plt.title(r"夹逼定理示例: $x^2 \sin(1/x)$ 被 $\pm x^2$ 夹逼")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
