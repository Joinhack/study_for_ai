## 16. 定积分

### 16.1. 基本思想
$\int_{a}^{b} f(x) \, dx$ 是由曲线$y=f(x)$,两条垂直线x=a和x=b,以及x轴所围成的有向面积

$$\int_{0}^{1} x \, dx = \frac{1}{2}$$  
$$\int_{0}^{2} x \, dx = 2$$
$$\int_{a}^{b} 1 \, dx = b-a$$

$$\int_{-\pi}^{\pi} sin(x) \, dx = 0$$

如果f(x)是奇函数那么$$\int_{-a}^{a} f(x) \, dx = 0$$

如果f(x)是偶函数那么$$\int_{-a}^{a} f(x) \, dx = 2\int_{0}^{a} f(x) \, dx $$

### 16.2. 定积分的定义
$$\int_{a}^{b} f(x) \, dx = \lim\limits_{mesh \to 0}\sum_{j=1}{n} f(c_j)(x_j - x_{j-1})$$
其中$a=x_0< x_1 <..< x_{n-1} < x_n =b $并且对于每一个j=1,...n都有$c_j$在$[x_{j-1}, x_j]$
### 16.3. 定积分的性质

$$\int_{b}^{a} f(x) \, dx = -\int_{a}^{b} f(x) \, dx $$

$$\int_{a}^{a} f(x) \, dx = 0 $$

$$\int_{a}^{b} f(x) \, dx = \int_{a}^{c} f(x) \, dx + \int_{c}^{b} f(x) \, dx $$

$$\int_{a}^{b} Cf(x) \, dx = C\int_{a}^{b} f(x) \, dx$$
$$\int_{a}^{b} (g(x) + f(x)) \, dx = \int_{a}^{b} g(x) \, dx + \int_{a}^{b} f(x) \, dx$$
$$\int_{a}^{b} (g(x) - f(x)) \, dx = \int_{a}^{b} g(x) \, dx - \int_{a}^{b} f(x) \, dx$$

### 16.4. 求面积

在函数f与g之间的面积 =  $\int_{a}^{b} |g(x) - f(x)| \, dx$

$\int_{A}^{B} f^{-1}(y) dy$是函数$y = f(x)$、直线y=A 和y=B以及y轴所围成的面积，如果函数的反函数是存在的。

### 16.5. 估算积分

如果对于在区间[a,b]有 $f(x)<=g(x)$那么就有$\int_{a}^{b} f(x) \, dx <= \int_{a}^{b} g(x) \, dx$

如果对于在区间[a,b]有 $m<=f(x)<=M$那么就有$m(b-a)\int_{a}^{b} f(x) \, dx <= M(b-a)$

### 16.6.积分的平均值和中值定理

函数f在区间[a,b]内的平均值 = $\frac{1}{b-a} \int_{a}^{b} f(x) \, dx $

积分中值定理：如果函数f在闭区间[a,b]上连续，那么在开区间(a,b)内总有一点c,满足 $f(c)=\frac{1}{b-a} \int_{a}^{b} f(x) \, dx $

###  17.1.用其他函数的积分来表示的函数

###  17.2. 微积分的第一基本定理

微积分的第一基本定理：如果函数f在闭区间[a,b]上是连续的，定义的F为$F(x) = \int_{a}^{x} f(t) \, dt, x \in [a,b] $则F在开区间(a,b)内是可导函数，而且$F'(x) = f(x)$,因此就有$$\frac{d}{dx}\int_{a}^{x} f(t) \, dt = f(x)$$

如果函数$F'(x) = f(x)$ 那么F(x)就是f(x)的不定积分记做$$\int f(x) \,  dx = F(x) + C$$

 反导函数：F的导数是$x^2$也就是说F是$x^2$的反导数，例子：$\int_{0}^{x} t^2 \, dt = \frac{x^3}{3} +C$

### 17.3.微积分的第二基本定理

微积分第二基本定理：如果函数f在闭区间[a,b]是连续的,F是f反导数（关于x）那么就有 $\int_{a}^{b} f(x) \, dx = F(b) - F(a) $ 也就$F(x) \Big|_a^b = F(b) - F(a)$

### 17.4.不定积分

$$\text{如果} \frac{d} {dx}F(x) = f(x) \text{那么} \int f(x)\, dx= F(x)$$

例子： $(\frac{x^3}{3} + C)' =  x^2$ 也就有$\int x^2 \, dx = \frac{x^3}{3} + C $
$\frac{d} {dx}sin(x) = cos(x)$ 也就有$\int cos(x) \, dx = sin(x) + C $


### 17.5.怎样解决问题：微积分的第一基本定理

$\frac{d}{dx}\int_{3}^{x}sin(t^2) \, dt$ 根据第一定律 将x换成t然后就有$\frac{d}{dx}\int_{3}^{x}sin(t^2) \, dt = sin(x^2)$

积分上下限的翻转$\int_{a}^{b}f(x) \, dx = -\int_{b}^{a}f(x) \, dx $
那么$\frac{d}{dx}\int_{x}^{7}t^3cos(tln(t)) \, dt = -\frac{d}{dx}\int_{7}^{x}t^3cos(tln(t)) \, dt = -x^3cos(xln(x))$


### 17.6.怎样解决问题：微积分的第二基本定理
如果$a\neq0$，这时$a+1\neq0$ 所以$\int x^a \, dx = \frac{x^{a+1}}{a+1} + C$ 

例子：$\frac{d}{dx} (x^4) = 4x^3$ 所以 $\int 4x^3\,dx = x^4 + C$ 因此根据第一定律$\int x^3\,dx = \frac{x^4}{4} + \frac{C}{4}$ 也就是 $\int x^3\,dx = \frac{x^4}{4} + C$

$$\frac{d}{dx}x^a = ax^{a-1}    \quad \quad  \int x^a\, dx = \frac{x^{a+1}}{a+1} +C $$
$$\frac{d}{dx}ln(x) = \frac{1}{x}    \quad \quad  \int \frac{1}{x} \, dx = ln|x| +C $$
$$\frac{d}{dx}e^x = e^x    \quad \quad  \int e^x \, dx = e^x +C $$
$$\frac{d}{dx}b^x = b^x\,ln(b)    \quad \quad  \int b^x \, dx = \frac{b^x}{ln(b)} +C $$
$$\frac{d}{dx}sin(x) = cos(x)    \quad \quad  \int cos(x) \, dx = sin(x) +C $$
$$\frac{d}{dx}cos(x) = -sin(x)    \quad \quad  \int sin(x) \, dx = -cos(x) +C $$
$$\frac{d}{dx}tan(x) = sec^2(x)    \quad \quad  \int sec^2(x) \, dx = tan(x) +C $$
$$\frac{d}{dx}sec(x) = sec(x)tan(x)    \quad \quad  \int sec(x)tan(x) \, dx = sec(x) +C $$
$$\frac{d}{dx}cot(x) = -csc^2(x)    \quad \quad  \int csc^2(x) \, dx = -cot(x) +C $$
$$\frac{d}{dx}csc(x) = -csc(x)cot(x)    \quad \quad  \int csc(x)cot(x) \, dx = -csc(x) +C $$
$$\frac{d}{dx}sin^{-1}(x) = \frac{1}{\sqrt{1-x^2}}    \quad \quad  \int \frac{1}{\sqrt{1-x^2}} \, dx = sin^{-1}(x) +C $$

$$\frac{d}{dx}tan^{-1}(x) = \frac{1}{\sqrt{1+x^2}}    \quad \quad  \int \frac{1}{\sqrt{1+x^2}} \, dx = sin^{-1}(x) +C $$

$$\frac{d}{dx}sec^{-1}(x) = \frac{1}{|x|\sqrt{x^2-1}}    \quad \quad  \int \frac{1}{|x|\sqrt{x^2-1}} \, dx = sec^{-1}(x) +C $$

$$\frac{d}{dx}sinh(x) = cosh(x)   \quad \quad  \int cosh(x) \, dx = sinh(x) +C $$
$$\frac{d}{dx}cosh(x) = sinh(x)   \quad \quad  \int sinh(x) \, dx = cosh(x) +C $$

### 18.1.换元法概述.


$$\int \frac{f'(x)}{f(x)} dx = \int \frac{df(x)}{f(x)} = \int \frac{dt}{t} = ln|t| + C = ln|f(x)| + C$$

### 18.1.2-18.1.3.如何换元&换元法的理论解释

$\int \frac{1}{sin^{-1}(x)\sqrt{1-x^2}} \, dx = \int \frac{1}{sin^{-1}(x)} \, d(sin^{-1}(x)) = \frac{1}{2} (sin^{-1}(x))^2 + C$

- 对于不定积分，用 t 和 dt 分别去表示带有 x 的表达式和dx，然后再求这个新的用 t 表达的积分，最后再换回到 x。
- 对于定积分，用t 和dt 分别去表示带有 x 的表达式和dx，并且也要把积分上下限换为与 t 相关的，这时计算这个新的积分（没有必要再回到以 x 为变量的状态）。当然也可以用另一个方法，那就是先把它看成不定积分计算结果，然后再把积分上下限分别代入求最后的结果。
